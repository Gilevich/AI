```
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from keras.datasets import boston_housing
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
```

Załadujemy zestaw danych.

```
(x_train, y_train), (x_test, y_test) = boston_housing.load_data()
```

Odejmujemy średnią i dzielimy przez odchylenie standardowe

```
# Średnia wartość
mean = x_train.mean(axis=0)
# Odchylenie standardowe
std = x_train.std(axis=0)
x_train -= mean
x_train /= std
x_test -= mean
x_test /= std
```

Tworzymy sieć
Warstwa wyjściowa z jednym neuronem liniowym - w przypadku zadanui regresji funkcja aktywacji nie jest używana.
Funkcja błędu to odchylenie standardowe. Metryka to średnie bezwzględne odchylenie.

```
model = Sequential()
model.add(Dense(128, activation='relu', input_shape=(x_train.shape[1],)))
model.add(Dense(1))
model.compile(optimizer='adadelta', loss='mse', metrics=['mae'])
Algorithms = np.array(['adadelta','adagrad','adam','rmsprop'])
Answers = np.empty([3,4])
```

```
k = 0;
for i in Algorithms:
    for j in range(3):
        print(i, j)
        model.fit(x_train, y_train, epochs=30, batch_size=1, verbose=2)
        mse, mae = model.evaluate(x_test, y_test, verbose=0)
        print('Erorr:', mae)
        pred = model.predict(x_test)
        print("Przewidywana wartość:", pred[1][0], ", właściwa cena:", y_test[1])
        Answers[j][k] = mae;
        model = Sequential()
        model.add(Dense(128, activation='relu', input_shape=(x_train.shape[1],)))
        model.add(Dense(1))
        model.compile(optimizer=i, loss='mse', metrics=['mae'])
    k+=1
```

Uczymy sieć


Ocena dokładności sieci

```
for i in range(4):
  print(Algorithms[i], np.around(Answers[:,i],3))
```
